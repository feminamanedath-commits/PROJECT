<h1>📊 Retail Analytics – Gaming Sales</h1>

<h2>Project Summary</h2>

This project applied retail analytics techniques to a gaming sales dataset, treating genres as product categories, consoles as sales channels, regions as markets, and critic scores as customer reviews. The goal was to uncover actionable insights into consumer preferences, platform performance, and market trends.

<h2>Key Insights</h2>

  <li><b>Category Performance</b>: Sports, Action, and Shooter genres dominate sales, functioning as “blockbuster categories,” while niche genres contribute marginally.</li>
  <li><b>Channel Benchmarking</b>: PlayStation 2, Xbox 360, and PlayStation 3 emerged as the most successful sales channels, reflecting the mid‑2000s console boom.</li>
  <li><b>Market Segmentation</b>: North America and PAL regions favor Sports and Shooter, while Japan leans toward Action and Role‑Playing — clear evidence of regional segmentation.</li>
  <li><b>Customer Reviews Impact</b>: Critic scores show a moderate correlation (~0.28) with sales, meaning product quality influences demand but is not the sole driver.</li>
  <li><b>Lifecycle Trends</b>: Sales rose steadily in the 1990s–2000s, peaked in the mid‑2000s to early 2010s, and declined post‑2015, reflecting market maturity and digital disruption.</li>
</ul>

<h2>Key Metrics</h2>

<ul>
  <li><b>Top categories by sales</b>: Sports (1187.5), Action (1125.9), Shooter (995.5)</li>
  <li><b>Top channels by sales</b>: PS2 (1027.8), X360 (859.8), PS3 (809.7)</li>
  <li><b>Regional correlation</b>: NA vs PAL ~0.91 (strong), NA vs JP ~0.21 (weak)</li>
  <li><b>Critic score correlation</b>: ~0.28 with total sales</li>
  <li><b>Peak demand period</b>: Mid‑2000s to early 2010s</li>
</ul>

<h2>Recommendations</h2>

  <li><b>Category Strategy</b>: Prioritize blockbuster genres while tailoring niche offerings to regional markets.</li>
  <li><b>Channel Focus</b>: Invest in dominant platforms (PlayStation, Xbox) while monitoring emerging channels.</li>
  <li><b>Regional Targeting</b>: Customize product portfolios — Sports/Shooter for Western markets, Role‑Playing/Adventure for Japan.</li>
  <li><b>Quality Emphasis</b>: Maintain strong critic scores to enhance visibility and competitiveness.</li>
  <li><b>Future Adaptation</b>: Address declining sales by embracing digital distribution, mobile gaming, and cross‑platform strategies.</li>
</ul>

<h2>Conclusion</h2>
<p>
This retail analytics study of gaming sales highlights how <b>product categories, sales channels, market segmentation, and customer reviews</b> shape industry performance. 
The dataset reveals a classic retail lifecycle: growth, peak, and decline. 
Companies that align category focus, channel strategy, and regional targeting — while adapting to digital transformation — will be best positioned to sustain growth in the evolving gaming market.
</p>
